```python
# TODO 06 - which Segment has the highest profit in 2018
filtered_df = df[df['Order Date'].dt.year == 2018]
filtered_df.groupby('Segment')['Profit'].sum()
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019

filtered_df = df[(df['Order Date'] >= '04-15-2019') & (df['Order Date'] <= '12-31-2019')]
State = filtered_df.groupby('State')['Sales'].sum()
State.sort_values(ascending=True).head(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
dfsales = df[((df['Region'] == 'West') | (df['Region'] == 'Central')) & (df['Order Date'].dt.year == 2019)]
total_sales_west = dfsales['Sales'].sum()

df2 = df[df['Order Date'].dt.year == 2019]
total_sales = df2['Sales'].sum()

(total_sales_west/total_sales)*100
```
